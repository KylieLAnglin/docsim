```python
import pandas as pd
import numpy as np
from scipy import spatial
```

```python
clean_filepath = "/Users/kylieleblancKylie/domino/docsim/data/clean/"
```

```python
docs = pd.read_csv(clean_filepath + 'corpus_text_noavatars.csv')
docs = docs.set_index('doc')
docs.sample(5)
```

```python
matrix = pd.read_csv(clean_filepath + 'doc_term_matrix_corpus_noavatars.csv')
matrix = matrix.set_index('doc')
matrix.sample(5)
```

```python
ideal_matrix = pd.read_csv(clean_filepath + 'doc_term_matrix_ideal_noavatars.csv')
ideal_matrix = ideal_matrix.set_index('doc')
ideal_matrix.sample()
```

```python
centroid = np.mean(ideal_matrix)
centroid
```

```python
def distance_to_ideal(df, semester, year):
    for maindoc in list(docs[(docs.year == year) & (docs.semester == semester)].index):
        pairwise_sim = []
        sim = 1 - spatial.distance.cosine(matrix.loc[maindoc], centroid)
        pairwise_sim.append(sim)
        average = sum(pairwise_sim)/len(pairwise_sim) # don't include relationship with self
        docs.at[maindoc, 'to_ideal'] = average
    return df
distance_to_ideal(docs, 'spring', '2018-19')
docs.sample(5)
```
