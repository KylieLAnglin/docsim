```python
import similarity
import operator
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
from scipy import spatial
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
from statistics import stdev
```

```python
fall2017_filepath = '/Users/kylieleblancKylie/domino/docsim/data/fall_2017/coaching/'
fall2018_filepath = '/Users/kylieleblancKylie/domino/docsim/data/fall_2018/coaching/'
spring2018_filepath = '/Users/kylieleblancKylie/domino/docsim/data/spring_2018/coaching/'
spring2019_filepath = '/Users/kylieleblancKylie/domino/docsim/data/spring_2019/coaching/'
```

# Extract text from documents

```python
fall2018_dict = similarity.make_text_dict(fall2018_filepath, '2018*docx')
spring2018_dict = similarity.make_text_dict(spring2018_filepath, '*docx')
spring2019_dict = similarity.make_text_dict(spring2019_filepath, '2019*')
full_dict = {}
full_dict.update(fall2018_dict)
full_dict.update(spring2019_dict)
```

```python
def create_df(textdict, year, semester):
    df = pd.DataFrame.from_dict(data = textdict, orient = 'index').reset_index()
    df = df.rename({'index': 'doc', 0: 'text'}, axis = 'columns')
    df['year'] = year
    df['semester'] = semester
    return df
fall2018 = create_df(fall2018_dict, 2018, 'fall')
spring2018 = create_df(spring2018_dict, 2018, 'spring')
spring2019 = create_df(spring2019_dict, 2019, 'spring')
```

```python
print(len(spring2018))
print(len(fall2018))
print(len(spring2019))
```

```python
df = fall2018.append(spring2018)
df = df.append(spring2019)
df
```

```python
print("First {} documents are from the Fall 2018.".format(len(fall2018_dict)))
print("Last {} documents are from the Spring 2019.".format(len(spring2019_dict)))
fall2018_start = 0
fall2018_end = len(fall2018_dict)
spring2019_start = len(fall2018_dict)
spring2019_end = len(full_dict)
print("Fall range: {} to {}".format(fall2018_start, fall2018_end))
print("Spring range: {} to {}".format(spring2019_start, spring2019_end))
```

# Create document term matrix for full copus

```python
corpus = list(full_dict.values())
vectorizer = CountVectorizer(token_pattern=r'\b[^\d\W]+\b')  # automatically tokenizes all words >2 characters
vectors = vectorizer.fit_transform(corpus)
print("Words extracted from corpus: ")
for word in vectorizer.get_feature_names()[0:200]:
    print(word, end = ", ")
```

```python
cosine_sim = cosine_similarity(vectors, dense_output=True)
print("Example cosine similarity matrix for one document: ")
np.set_printoptions(precision=2)
cosine_sim[0]
```

```python
fall2018_sim_metrics = {}
for doc, loc in zip(fall2018_dict, range(0, len(fall2018_dict))):
    fall2018_sim_metrics[doc] = ((sum(cosine_sim[loc][0:len(fall2018_dict)]) - 1)/(len(fall2018_dict) - 1), 
                             cosine_sim[loc][49:91].mean())
fall2018_within = [x[0] for x in fall2018_sim_metrics.values()]
fall2018_within_mean = sum(fall2018_within)/len(fall2018_within)
fall2018_between = [x[1] for x in fall2018_sim_metrics.values()]
fall2018_between_mean = sum(fall2018_between)/len(fall2018_between)
print("Average Document Similarity")
print("Fall within: ", round(fall2018_within_mean, ndigits = 3), "Fall Between: ", round(fall2018_between_mean, ndigits = 3))
print("            [", round(stdev(fall2018_within), ndigits = 3), "]            [", round(stdev(fall2018_between), ndigits = 3), "]")
```

```python
spring2019_sim_metrics = {}
for doc, loc in zip(spring2019_dict, range(49, 91)):
    spring2019_sim_metrics[doc] = ((sum(cosine_sim[loc][49:91]) - 1)/(42 - 1), 
                               cosine_sim[loc][0:49].mean())
spring2019_within = [x[0] for x in spring2019_sim_metrics.values()]
spring2019_within_mean = sum(spring2019_within)/len(spring2019_within)
spring2019_between = [x[1] for x in spring2019_sim_metrics.values()]
spring2019_between_mean = sum(spring2019_between)/len(spring2019_between)
print("Spring within: ", round(spring2019_within_mean, ndigits = 3), "Spring Between: ", round(spring2019_between_mean, ndigits = 3))
print("              [", round(stdev(spring2019_within), ndigits = 3), "]            [", round(stdev(spring2019_between), ndigits = 3), "]")
```

```python
bins = np.linspace(0.6, 1, num = 40)
plt.hist(fall2018_within, bins, color = "blue", label = "Fall Within", alpha = .5)
#plt.hist(spring_within, bins, color = "green", label = "Spring Within", alpha = .5)
plt.hist(fall2018_between, bins, color = "orange", label = "Fall to Spring", alpha = .5)
#plt.hist(spring_between, bins, color = "yellow", label = "Spring Between", alpha = .5)
plt.legend()
plt.title("Fall Coaching Similarity")
plt.show()
```

```python
plt.hist(spring2019_within, bins, color = "green", label = "Spring Within", alpha = .5)
plt.hist(spring2019_between, bins, color = "yellow", label = "Spring to Fall", alpha = .5)
plt.legend()
plt.title("Spring Coaching Similarity")
plt.show()
```
