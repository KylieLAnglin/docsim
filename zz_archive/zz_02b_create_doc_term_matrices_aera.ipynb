```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import spacy
```

```python
clean_filepath = "/Users/kylieleblancKylie/domino/docsim/data/clean/aera/"
```

```python
df_open = pd.read_csv(clean_filepath + 'text_aera_open.csv')
df_open = df_open.set_index('doc')
df_open['special'] = 0
df_open.sample(5)
```

```python
df_special = pd.read_csv(clean_filepath + 'text_aera_special.csv')
df_special = df_special.set_index('doc')
df_special['special'] = 1
df_special.sample(3)
```

```python
df = df_open.append(df_special, sort = True)
df.sample(5)
```

# Doc-Term Matrix

```python
docs = list(df.text)
vec = CountVectorizer()
X = vec.fit_transform(docs)
matrix = pd.DataFrame(X.toarray(), columns=vec.get_feature_names(), index = df.index)
print('Number of words: ', len(matrix.columns))
matrix.sample()
```

```python
matrix_special = matrix[df.special == 1]
matrix_special.to_csv(clean_filepath + 'matrix_special.csv')
matrix_special.sample()
```

```python
matrix_open = matrix[df.special == 0]
matrix_open.to_csv(clean_filepath + 'matrix_open.csv')
matrix_open.sample()
```

# Weight

```python
docs = list(df.text)
vec = TfidfVectorizer()
X = vec.fit_transform(docs)
matrix_wgt = pd.DataFrame(X.toarray(), columns=vec.get_feature_names(), index = df.index)
print('Number of words: ', len(matrix_wgt.columns))
matrix_wgt.sample()

matrix_special_wgt = matrix_wgt[df.special == 1]
matrix_special_wgt.to_csv(clean_filepath + 'matrix_special_wgt.csv')
matrix_special_wgt.sample()


matrix_open_wgt = matrix_wgt[df.special == 0]
matrix_open_wgt.to_csv(clean_filepath + 'matrix_open_wgt.csv')
matrix_open_wgt.sample()
```

# Remove Stop Words (no weighting)

```python
stop_words = list(stopwords.words('english'))
stop_words.append('like')
stop_words.append('um')
stop_words.append('uhm')
stop_words.append('-pron-')
stop_words
```

```python
docs = list(df.text)
vec = CountVectorizer(stop_words = stop_words)
X = vec.fit_transform(docs)
matrix_stop = pd.DataFrame(X.toarray(), columns=vec.get_feature_names(), index = df.index)
print('Number of words: ', len(matrix_stop.columns))
matrix_stop.sample()
```

```python
matrix_special_stop = matrix_stop[df.special == 1]
matrix_special_stop.to_csv(clean_filepath + 'matrix_special_stop.csv')
matrix_special_stop.sample()
```

```python
matrix_open_stop = matrix_stop[df.special == 0]
matrix_open_stop.to_csv(clean_filepath + 'matrix_open_stop.csv')
matrix_open_stop.sample()
```

# Stop and weight

```python
docs = list(df.text)
vec = TfidfVectorizer(stop_words = stop_words)
X = vec.fit_transform(docs)
matrix_stop_wgt = pd.DataFrame(X.toarray(), columns=vec.get_feature_names(), index = df.index)
print('Number of words: ', len(matrix_stop_wgt.columns))
matrix_stop_wgt.sample()

matrix_special_stop_wgt = matrix_stop_wgt[df.special == 1]
matrix_special_stop_wgt.to_csv(clean_filepath + 'matrix_special_stop_wgt.csv')
matrix_special_stop_wgt.sample()


matrix_open_stop_wgt = matrix_stop_wgt[df.special == 0]
matrix_open_stop_wgt.to_csv(clean_filepath + 'matrix_open_stop_wgt.csv')
matrix_open_stop_wgt.sample()
```

# Stem

```python
# Initialize spacy 'en' model, keeping only tagger component needed for lemmatization
nlp = spacy.load('en', disable=['parser', 'ner'])

sentence = "The striped bats are hanging on their feet for best singing. My favorite color is green. Did you know that?"

# Parse the sentence using the loaded 'en' model object `nlp`
doc = nlp(sentence)

# Extract the lemma for each token and join
" ".join([token.lemma_ for token in doc])
#> 'the strip bat be hang on -PRON- foot for good'
```

```python
stems = [" ".join([token.lemma_ for token in nlp(text)]) for text in df.text]
df['text'] = stems
df.sample(5)
```

```python
docs = list(df.text)
vec = CountVectorizer(stop_words = stop_words)
X = vec.fit_transform(docs)
matrix_stop_stem = pd.DataFrame(X.toarray(), columns=vec.get_feature_names(), index = df.index)
print('Number of words: ', len(matrix_stop_stem.columns))
matrix_stop_stem.sample()
```

```python
matrix_special_stop_stem = matrix_stop_stem[matrix.special == 1]
matrix_special_stop_stem.to_csv(clean_filepath + 'matrix_special_stop_stem.csv')
matrix_special_stop_stem.sample()
```

```python
matrix_open_stop_stem = matrix_stop_stem[matrix.special == 0]
matrix_open_stop_stem.to_csv(clean_filepath + 'matrix_open_stop_stem.csv')
matrix_open_stop_stem.sample()
```
